```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
picher_file_path = 'data/picher_stats_2017.csv'
batter_file_path = 'data/batter_stats_2017.csv'
picher = pd.read_csv(picher_file_path)
batter = pd.read_csv(batter_file_path)
```

```python
batters=batter[["팀명","연봉(2017)"]]#타자 팀명 연봉
pichers=picher[["팀명","연봉(2017)"]]#투수 팀명 연봉
total=pd.merge(batters,pichers)#타자 투수 데이터 합
han=total.loc[total["팀명"]=="한화"]
sk=total.loc[total["팀명"]=="SK"]

tTestResult = stats.ttest_ind(han["연봉(2017)"],sk["연봉(2017)"])
tTestResultDiffVar = stats.ttest_ind(han["연봉(2017)"],sk["연봉(2017)"], equal_var=False)

print("The t-statistic and p-value assuming equal variances is %.3f and %.3f." % tTestResult)
print("The t-statistic and p-value not assuming equal variances is %.3f and %.3f" % tTestResultDiffVar)
```
